```python
import pandas as pd
from lavague.core.new_evaluator import RetrieverEvalutor

dataset = pd.read_csv("../../lavague-labelizer/data_for_retriever_eval.csv")
evaluator = RetrieverEvalutor()
dataset.head()
```

```python
from lavague.core.retrievers import BM25HtmlRetriever

bm25retriever = BM25HtmlRetriever()
bm25_results = evaluator.evaluate(bm25retriever, dataset)
bm25_results.head()
```

```python
from lavague.core.retrievers import get_default_retriever
from lavague.drivers.selenium import SeleniumDriver
import os

os.environ["OPENAI_API_KEY"] = None
driver = SeleniumDriver()
default_retriever = get_default_retriever(driver)
default_results = evaluator.evaluate(default_retriever, dataset)
default_results.head()
```
